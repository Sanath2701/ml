```python
# 📦 Install (for Colab)
!pip install openpyxl ipywidgets matplotlib pandas

# 📁 Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

# 📥 Step 2: Load the Excel file
file_path = "/content/ROWA_Labcar_Validation_Plan_1.0 NEW.xlsx"
sheet_name = "ROWA_EV_Validation_Plan"

# Load DataFrame with headers at row 1
df = pd.read_excel(file_path, sheet_name=sheet_name, header=1)

# 🧹 Step 3: Clean the DataFrame
df = df.loc[:, df.columns.notna()]  # Drop unnamed columns
df = df[['Features', 'No. of Test cases', 'PASS', 'FAIL', 'NOT TESTED', 'NA']]
df = df.dropna(subset=['Features'])  # Drop rows with no Features
df['No. of Test cases'] = pd.to_numeric(df['No. of Test cases'], errors='coerce').fillna(0).astype(int)

# 🎯 Step 4: TOTAL PIE CHART - Show distribution of test cases across all features
def show_total_distribution():
    total_cases_by_feature = df.groupby('Features')['No. of Test cases'].sum()
    labels = total_cases_by_feature.index.tolist()
    values = total_cases_by_feature.values.tolist()

    plt.figure(figsize=(8, 8))
    plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
    plt.title("📊 Total Test Case Distribution by Feature", fontsize=14)
    plt.tight_layout()
    plt.show()

# 🧠 Step 5: Dashboard for individual feature
def update_feature_dashboard(feature):
    filtered = df[df['Features'].str.contains(feature, case=False, na=False)]

    if filtered.empty:
        print(f"No data found for '{feature}'")
        return

    total_cases = filtered['No. of Test cases'].sum()
    avg_cases = filtered['No. of Test cases'].mean()
    count_cases = filtered.shape[0]

    print(f"📌 Feature: {feature}")
    print(f"✅ Total Test Cases: {total_cases}")
    print(f"📊 Average per Entry: {avg_cases:.2f}")
    print(f"🔢 Entry Count: {count_cases}")

    # 🥧 Pie 1: Result Distribution
    pass_count = filtered['PASS'].sum()
    fail_count = filtered['FAIL'].sum()
    not_tested_count = filtered['NOT TESTED'].sum()
    na_count = filtered['NA'].sum()

    result_labels = ['PASS', 'FAIL', 'NOT TESTED', 'NA']
    result_values = [pass_count, fail_count, not_tested_count, na_count]

    # 🥧 Pie 2: Sub-feature Test Case Count
    sub_labels = filtered['Features'].tolist()
    sub_values = filtered['No. of Test cases'].tolist()

    fig, axs = plt.subplots(1, 2, figsize=(14, 6))

    axs[0].pie(result_values, labels=result_labels, autopct='%1.1f%%', startangle=90,
               colors=['green', 'red', 'orange', 'gray'])
    axs[0].set_title('✅ Test Result Distribution')

    axs[1].pie(sub_values, labels=sub_labels, autopct='%1.1f%%', startangle=90,
               textprops={'fontsize': 8})
    axs[1].set_title('📌 Entry-wise Test Case Count')

    plt.tight_layout()
    plt.show()

# 🎬 Step 6: Run the dashboard
print("📊 Showing Overall Test Case Distribution...")
show_total_distribution()

# 🎛️ Step 7: Dropdown for feature selection
feature_options = sorted(df['Features'].dropna().unique().tolist())
interact(update_feature_dashboard, feature=widgets.Dropdown(options=feature_options, description="🔧 Feature:"));
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact, widgets

file_path = "/content/ROWA_Labcar_Validation_Plan_1.0 NEW.xlsx"
sheet_name = "ROWA_EV_Validation_Plan"

df = pd.read_excel(file_path, sheet_name=sheet_name, header=13)
df = df.loc[:, df.columns.notna()]
df = df[['Features', 'No. of Test cases', 'PASS', 'FAIL', 'NOT TESTED', 'NA']]
df = df.dropna(subset=['Features'])


df['No. of Test cases'] = pd.to_numeric(df['No. of Test cases'], errors='coerce').fillna(0).astype(int)
for col in ['PASS', 'FAIL', 'NOT TESTED', 'NA']:
    df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0).astype(int)


def classify_component(feature):
    feature = feature.lower()
    if "airbag" in feature:
        return "Airbag"
    elif "bcm" in feature or "body control" in feature:
        return "BCM"
    elif "lamp" in feature or "lighting" in feature:
        return "Lighting"
    elif "wiper" in feature:
        return "Wiper"
    elif "door" in feature or "tailgate" in feature:
        return "Doors & Tailgate"
    elif "cluster" in feature:
        return "Instrument Cluster"
    elif "seat" in feature or "squab" in feature:
        return "Seat System"
    elif "window" in feature:
        return "Window"
    else:
        return "Others"

df["Component"] = df["Features"].apply(classify_component)

def show_component_dashboard(component):
    comp_df = df[df['Component'] == component]

    print(f" Component: {component}")
    print(f" Total Test Cases: {comp_df['No. of Test cases'].sum()}")
    display(comp_df[['Features', 'No. of Test cases']])

    plt.figure(figsize=(7, 5))
    plt.pie(comp_df['No. of Test cases'],
            labels=comp_df['Features'],
            autopct='%1.1f%%',
            startangle=90,
            textprops={'fontsize': 8})
    plt.title(f"Component-wise Test Case Split - {component}")
    plt.show()


def show_feature_pie(feature):
    filtered = df[df['Features'] == feature]
    if filtered.empty:
        print(f"No data for {feature}")
        return

    print(f" Feature: {feature}")
    print(filtered[['No. of Test cases', 'PASS', 'FAIL', 'NOT TESTED', 'NA']])

    values = filtered[['PASS', 'FAIL', 'NOT TESTED', 'NA']].values[0]
    total = sum(values)

    if total == 0:
        print(" No test result data to display (all values are 0).")
        return

    labels = ['PASS', 'FAIL', 'NOT TESTED', 'NA']
    colors = ['green', 'red', 'orange', 'gray']

    plt.figure(figsize=(6, 6))
    plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors)
    plt.title("Test Result Distribution")
    plt.show()



print("Select a Component:")
interact(show_component_dashboard, component=sorted(df['Component'].unique()))
print("Select a Feature:")
interact(show_feature_pie, feature=sorted(df['Features'].unique()))
```
